```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df["job_skills"] = df["job_skills"].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
df_DA = df[df["job_title_short"] == "Data Analyst"].copy()
```

```python
df_DA
```

```python
df_DA["job_posted_month_no"] = df_DA["job_posted_date"].dt.month
df_DA
```

```python
df_DA_explode = df_DA.explode("job_skills").copy()
```

```python
df_DA_explode
```

```python
df_DA_explode.pivot_table(index="job_posted_month_no", columns="job_skills", aggfunc="size")
```

```python
df_DA_explode.pivot_table(index="job_posted_month_no", columns="job_skills", aggfunc="size", fill_value=0)
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index="job_posted_month_no", columns="job_skills", aggfunc="size", fill_value=0)
```

```python
df_DA_pivot.plot(kind="line")
```

```python
df_DA_pivot.loc["Total"] = df_DA_pivot.sum()
df_DA_pivot
```

```python
df_DA_pivot.sort_values(axis=1, by="Total", ascending=False)
```

```python
df_DA_pivot_ordered = df_DA_pivot.sort_values(axis=1, by="Total", ascending=False).copy()
```

```python
df_DA_pivot_ordered
```

```python
df_DA_pivot_completed = df_DA_pivot_ordered.drop("Total").copy()
df_DA_pivot_completed
```

```python
df_DA_pivot_completed.iloc[:, 0:5]
```

```python
df_DA_pivot_completed.iloc[:, 0:5].plot(kind="line")
plt.title("Top 5 Skills for Data Analysts per Month")
plt.ylabel("Count")
plt.xlabel("")
plt.show()
```
