### Index Management

```python
# Import Libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")  # import data set form https://huggingface.co/datasets/lukebarousse/data_jobs
df = dataset["train"].to_pandas()  # change datatype
```

```python
# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df.sample(5)
```

```python
df.index
```

```python
df.index.name = "job_index"
```

```python
df.head()
```

```python
df.index.dtype
```

### Three Methods to cover 
- `df.reset_index()`
- `df.set_index()`
- `df.sort_index()`

### reset_index()

```python
df_usa = df[df["job_country"] == "United States"]
```

```python
df_usa
```

```python
df_usa.reset_index(inplace=True)
```

```python
df_usa
```

## set_index()

```python
df_usa.set_index("job_index", inplace=True)
```

```python
df_usa
```

### sort_index()

```python
median_pivot = df_usa.pivot_table(values="salary_year_avg", index="job_title_short", aggfunc=["median", "min", "max"])
```

```python
median_pivot
```

```python
median_pivot.sort_values(by=("median", "salary_year_avg"), ascending=False)
```

```python
median_pivot.sort_index(inplace=True)
```
