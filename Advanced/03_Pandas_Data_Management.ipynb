#### Why using copy()?

```python
# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")  # import data set form https://huggingface.co/datasets/lukebarousse/data_jobs
df_original = dataset["train"].to_pandas()  # change datatype
```

```python
df_original
```

```python
df_altered = df_original  # it's not how to create copy!!!
```

```python
df_altered.loc[:5, "salary_year_avg"]
```

```python
median_salary = df_altered["salary_year_avg"].median()
median_salary
```

```python
df_altered["salary_year_avg"] = df_altered["salary_year_avg"].fillna(median_salary)
```

```python
df_altered.loc[:5, "salary_year_avg"]
```

```python
df_original.loc[:5, "salary_year_avg"]  # we updated also original one 
```

```python
# same id --> that's why we need to use copy method and not use var1 = var2
print(id(df_altered))
id(df_original)
```

```python
id(df_altered) == id(df_original)
```

#### copy() usage

```python
# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")  # import data set form https://huggingface.co/datasets/lukebarousse/data_jobs
df_original = dataset["train"].to_pandas()  # change datatype
```

```python
df_altered = df_original.copy()
```

```python
print(id(df_altered))
id(df_original)
```

```python
id(df_altered) == id(df_original)  # diffrent id --> we can now update one without changing secound one
```

```python
median_salary = df_altered["salary_year_avg"].median()
median_salary
```

```python
df_altered["salary_year_avg"] = df_altered["salary_year_avg"].fillna(median_salary)
```

```python
df_altered.loc[0:5, "salary_year_avg"]
```

```python
df_original.loc[0:5, "salary_year_avg"]
```
