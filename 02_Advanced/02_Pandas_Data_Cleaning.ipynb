```python
# Import Libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")  # import data set form https://huggingface.co/datasets/lukebarousse/data_jobs
df = dataset["train"].to_pandas()  # change datatype
```

```python
# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df.loc[:10, "salary_year_avg":"salary_hour_avg"]
```

```python
df["salary_year_avg"].median()
```

```python
median_salary_year = df["salary_year_avg"].median()
```

```python
df["salary_hour_avg"].median()
```

```python
median_salary_hour = df["salary_hour_avg"].median()
```

#### fillna method

```python
df_filled = df
```

```python
df_filled["salary_year_avg"].fillna(median_salary_year)
```

```python
df_filled["salary_year_avg"] = df_filled["salary_year_avg"].fillna(median_salary_year)
```

```python
df_filled["salary_year_avg"]
```

```python
df_filled["salary_hour_avg"] = df_filled["salary_hour_avg"].fillna(median_salary_hour)
```

```python
df_filled["salary_hour_avg"]
```

```python
df_filled.loc[0:10, "salary_year_avg":"salary_hour_avg"]
```

#### Drop duplicates

```python
df_unique = df_filled
```

```python
df_unique.drop_duplicates()
```

```python
df_unique = df_unique.drop_duplicates()  # checing duplicates on all col
```

```python
print("Length of original df:      ", len(df_filled))
print("Length of drop duplicate df:", len(df_unique))
print("Rows Dropped:               ", len(df_filled) - len(df_unique))
```

```python
df_unique = df_unique.drop_duplicates(subset=["job_title", "company_name"])  # checing duplicates only in two col
```

```python
print("Length of original df:      ", len(df_filled))
print("Length of drop duplicate df:", len(df_unique))
print("Rows Dropped:               ", len(df_filled) - len(df_unique))
```
