## Exercise - Analyzing Data Science Job Demand

```python
# Import Libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")  # import data set form https://huggingface.co/datasets/lukebarousse/data_jobs
df = dataset["train"].to_pandas()  # change datatype
```

```python
# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df_US = df[df["job_country"] == "United States"].copy()
```

```python
df_US
```

```python
df_US["job_posted_month"] = df_US["job_posted_date"].dt.strftime("%B")  # documentation for .dt !!
```

```python
df_US
```

```python
df_US.pivot_table(index="job_posted_month", columns="job_title_short", aggfunc="size")
```

```python
df_US_pivot = df_US.pivot_table(index="job_posted_month", columns="job_title_short", aggfunc="size")
```

```python
df_US_pivot.reset_index(inplace=True)
```

```python
df_US_pivot
```

```python
df_US_pivot["month_no"] = pd.to_datetime(df_US_pivot["job_posted_month"], format="%B").dt.month
```

```python
df_US_pivot["month_no"]
```

```python
df_US_pivot
```

```python
df_US_pivot.sort_values("month_no", inplace=True)
```

```python
df_US_pivot
```

```python
df_US_pivot.set_index("job_posted_month", inplace=True)
```

```python
df_US_pivot.drop(columns="month_no", inplace=True)
```

```python
df_US_pivot
```

```python
df_US_pivot.plot(kind="line")
```

```python
top_3 = df_US["job_title_short"].value_counts().head(3)
```

```python
top_3
```

```python
top_3.index
```

```python
top_3.index.tolist()
```

```python
top_3 = top_3.index.tolist()
```

```python
top_3
```

```python
df_US_pivot
```

```python
df_US_pivot[top_3].plot(kind="line")
plt.title("Monthly Job Postings for Top Data Jobs in the US")
plt.xlabel("2023")
plt.ylabel("Job Count")
plt.legend()
plt.show()
```
