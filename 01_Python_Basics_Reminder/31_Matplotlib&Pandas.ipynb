```python
# Import Libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")  # import data set form https://huggingface.co/datasets/lukebarousse/data_jobs
df = dataset["train"].to_pandas()  # change datatype
```

```python
# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
job_counts = df.job_title_short.value_counts()
```

```python
type(job_counts)  # Series
```

```python
job_counts.plot(kind="bar")
```

```python
job_counts.plot(kind="bar")
plt.title("Posting by Job Title")
plt.ylabel("Count of Job Postings")
plt.xlabel("")
plt.xticks(rotation=45, ha="right") 
plt.show()
```

```python
type(df)  # DataFrame
```

```python
df.dropna(subset=["salary_year_avg"], inplace=True)
```

```python
# aggregating by month
df["job_posted_month"] = df["job_posted_date"].dt.month  # new col with months
```

```python
jobs_march = df[df["job_posted_month"] == 3]
```

```python
jobs_march
```

```python
jobs_march.plot(x="job_posted_date", y="salary_year_avg", kind="scatter")  # it's possible to create plots to Data Frames but in this particual example we have 2204 rows of data
plt.title("Job Posting by Date")
plt.ylabel("Salary Year Avg")
plt.xlabel("")
plt.xticks(rotation=45, ha="right") 
plt.show()
```
