#### NumPy faster statistics

```python
# create a list of 1,000,000 salaries ranging form 50,000 to 150,000

import random
salary_list = [random.randint(50000, 150000) for _ in range(10_000_000)]
salary_list
```

```python
len(salary_list)
```

```python
import statistics
```

```python
%%timeit

statistics.median(salary_list)
```

```python
import numpy as np
```

```python
%%timeit
np.median(salary_list)
```

```python
# Job titles
job_titles = np.array(["Data Analyst", "Data Scientist", "Data Engineer", "Machine Learning Engineer", "AI Engineer"])

# Base salaries
base_salaries = np.array([60000, 80000, 75000, 90000, np.nan])  # np.nan --> Null which is float data types and we can use it to mathematical operations!!!

# Bonus rates
bonus_rates = np.array([0.05, 0.1, 0.08, 0.12, np.nan])
```

```python
total_salaries = base_salaries * (1 + bonus_rates)
total_salaries
```

```python
np.nanmean(total_salaries)  # if we have nan value we have to use .nan function
```
