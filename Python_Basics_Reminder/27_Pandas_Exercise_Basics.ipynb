#### Analize Poland job market by looking key statistics

```python
import pandas as pd
from datasets import load_dataset
```

```python
# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")  # import data set form https://huggingface.co/datasets/lukebarousse/data_jobs
df = dataset["train"].to_pandas()  # change datatype
```

```python
df["job_country"].value_counts().head(20)
```

```python
df["job_country"].isin(["Poland"]).any()
```

```python
df["job_country"].isin(["North Korea"]).any()
```

```python
pl_jobs = df[df["job_country"] == "Poland"]
```

```python
pl_jobs
```

```python
pl_jobs = pl_jobs[pl_jobs["salary_year_avg"].notna()]
```

```python
pl_jobs
```

```python
pl_jobs.groupby("job_title_short")["salary_year_avg"].agg(["count","median", "min", "max"]).sort_values(by="median", ascending=False)
```
