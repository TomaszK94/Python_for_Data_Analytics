```python
# Import Libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")  # import data set form https://huggingface.co/datasets/lukebarousse/data_jobs
df = dataset["train"].to_pandas()  # change datatype
```

```python
# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

### Bar Chart - Counts of Job Postings

```python
job_counts = df.job_title_short.value_counts()
```

```python
job_counts
```

```python
plt.bar(job_counts.index, job_counts)  # need to be modified --> labels and more
plt.show()
```

```python
plt.bar(job_counts.index, job_counts) 
plt.title("Posting by Job Title")
plt.ylabel("Count of Job Postings")
plt.xticks(rotation=45, ha="right") 
plt.show()
```
