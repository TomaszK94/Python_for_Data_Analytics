## Median Salary vs Data Science Jobs

```python
# Import Libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")  # import data set form https://huggingface.co/datasets/lukebarousse/data_jobs
df = dataset["train"].to_pandas()  # change datatype
```

```python
# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df.groupby("job_title_short")["salary_year_avg"].median().sort_values()
```

```python
job_salary = df.groupby("job_title_short")["salary_year_avg"].median().sort_values()
```

```python
job_salary.plot(kind="barh")
plt.xlabel("Year Salary ($USD)")
plt.ylabel("")
plt.title("Median Salary by Job Title")
plt.show()
```
